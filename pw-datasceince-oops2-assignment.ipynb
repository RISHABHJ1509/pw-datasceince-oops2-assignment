```python
#12. Explain the use of the `super()` function in Python constructors. Provide an example.
'''
ans.In Python, the super() function is used to call a method from a parent class. When working with constructors, 
super() is often used to invoke the constructor of the parent class, allowing you to initialize the attributes defined
in the parent class before handling the specific initialization in the child class.'''

class ParentClass:
    def __init__(self, param1, param2):
        self.param1 = param1
        self.param2 = param2

    def display_params(self):
        print("Parent Class - Param1:", self.param1)
        print("Parent Class - Param2:", self.param2)


class ChildClass(ParentClass):
    def __init__(self, param1, param2, param3):
        # Call the constructor of the parent class using super()
        super().__init__(param1, param2)

        # Initialize the additional parameter specific to the child class
        self.param3 = param3

    def display_params(self):
        # Call the display_params method of the parent class using super()
        super().display_params()
        print("Child Class - Param3:", self.param3)


# Creating an object of the ChildClass
child_obj = ChildClass("A", "B", "C")

# Calling the display_params method of the ChildClass
child_obj.display_params()
```

```python
#16. How do you prevent a class from having multiple instances by using constructors in Python? Provide anexample.
'''
n Python, you can prevent a class from having multiple instances by using a class variable to track whether an 
instance has already been created. If an instance already exists, the constructor can raise an exception or take 
some other action to indicate that further instances are not allowed.'''

class SingletonClass:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(SingletonClass, cls).__new__(cls)
            return cls._instance
        else:
            raise Exception("SingletonClass can only have one instance")

    def __init__(self, data):
        if not hasattr(self, 'initialized'):
            self.data = data
            self.initialized = True

# Creating instances
try:
    obj1 = SingletonClass("Instance 1 Data")
    print(obj1.data)

    obj2 = SingletonClass("Instance 2 Data")  # This will raise an exception
except Exception as e:
    print(e)
```

```python
#19. Explain the use of constructor chaining in Python. Provide a practical example.
'''
ans.Constructor chaining in Python refers to the concept of one constructor calling another constructor within 
the same class. This allows you to reuse code and avoid duplicating initialization logic. In Python, constructor 
chaining is achieved using the super() function, which provides a way to call a method in a superclass.'''

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Student(Person):
    def __init__(self, name, age, student_id):
        # Calling the constructor of the superclass (Person)
        super().__init__(name, age)
        self.student_id = student_id

    def display_info(self):
        # Extending the display_info method of the superclass
        super().display_info()
        print(f"Student ID: {self.student_id}")

# Example usage:
student = Student("Alice", 20, "S12345")
student.display_info()
```

```python
#5. How can you access the methods and attributes of a parent class from a child class in Python? Give anexample.
'''
ans.In Python, you can access methods and attributes of a parent class from a child class using the super() function. 
The super() function provides a way to invoke methods and access attributes from the parent class within the child class.'''

class Vehicle:
    def __init__(self, color):
        self.color = color

    def start(self):
        return "Starting the vehicle..."
    
    
class Car(Vehicle):
    def __init__(self, color, brand):
        super().__init__(color)
        self.brand = brand

    def drive(self):
        return "Driving the car..."

    def start_and_drive(self):
        start_message = super().start()  # Accessing start() from parent class
        drive_message = self.drive()
        return f"{start_message} {drive_message}"
    
    
# Creating a Car instance
my_car = Car(color="Blue", brand="Toyota")

# Accessing attributes of Vehicle and Car
print(my_car.color) 
print(my_car.brand)  

# Accessing methods from Vehicle and Car
print(my_car.start())         
print(my_car.drive())          
print(my_car.start_and_drive())  
```

```python
'''12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an
example using the `abc` module.'''

'''
Abstract Base Classes (ABCs) in Python, available in the abc module, serve as a blueprint for other classes and
enforce specific methods to be implemented by their subclasses. They are a way of defining interfaces without
necessarily implementing all the details of those interfaces.'''

from abc import ABC, abstractmethod

class Shape(ABC):  # Inheriting from ABC
    @abstractmethod
    def area(self):
        pass  # Abstract method to be overridden
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
circle = Circle(radius=5)
rectangle = Rectangle(width=4, height=6)

print(f"Area of the circle: {circle.area():.2f}")        
print(f"Area of the rectangle: {rectangle.area()}")  
```

```python
#2. Describe the key principles of encapsulation, including access control and data hiding.
'''

 Encapsulation revolves around two key principles: access control and data hiding.

1. Access Control:
Encapsulation allows you to control the access to certain components (attributes and methods) of a class. 
In most object-oriented programming languages, there are three types of access control:

Public: These attributes and methods are accessible from anywhere. They can be freely accessed and modified 
from outside the class.

Private: Private attributes and methods are only accessible within the class that defines them. In many languages
like Python, this is achieved by using a naming convention (e.g., prefixing attributes/methods with double underscores __)
to make them less accessible from outside the class.

Protected: Protected attributes and methods are accessible within the class itself and its subclasses (derived classes). 
In some languages, this is denoted by a single underscore _ convention, but in Python, it's more of a naming convention 
to indicate that it should be treated as protected.

2. Data Hiding:
Data hiding is the concept of hiding the internal state of an object from the outside world and providing controlled 
access to it. This is done to ensure that the object's state is accessed and modified only through well-defined interfaces
(methods) provided by the class.

Private Data: By making certain attributes private, their values can only be modified or accessed through the class's
methods, allowing the class to have control over how its data is manipulated.

Encapsulation of Behavior: Encapsulation not only hides data but also encapsulates behavior by bundling the methods that
manipulate that data within the same class. This ensures that the object's data is modified in a controlled manner,
preventing unexpected changes that could lead to errors or bugs.'''
```

```python
#3. How can you achieve encapsulation in Python classes? Provide an example.

#ans.In Python, encapsulation can be achieved using access modifiers and conventions rather than strict enforcement
class Car:
    def __init__(self, make, model, year):
        self.__make = make  # Private attribute
        self.__model = model  # Private attribute
        self.__year = year  # Private attribute
        self._fuel = 100  # Protected attribute

    # Getter methods to access private attributes
    def get_make(self):
        return self.__make

    def get_model(self):
        return self.__model

    def get_year(self):
        return self.__year

    # Setter method to modify private attribute
    def set_year(self, year):
        self.__year = year

    # Method to drive the car and update fuel
    def drive(self, distance):
        if distance * 10 > self._fuel:
            print("Not enough fuel!")
        else:
            self._fuel -= distance * 10
            print(f"Car drove {distance} miles. Fuel remaining: {self._fuel}")


# Creating an instance of the Car class
my_car = Car("Toyota", "Corolla", 2020)

# Accessing attributes using getter methods
print(f"Car Make: {my_car.get_make()}")  # Accessing private attribute
print(f"Car Model: {my_car.get_model()}")  # Accessing private attribute
print(f"Car Year: {my_car.get_year()}")  # Accessing private attribute

# Trying to modify a private attribute directly (not recommended)
# This is not the standard practice as private attributes should be accessed via setter methods
# For demonstration purposes only
my_car.__year = 2022

# Accessing and modifying private attribute using setter method
my_car.set_year(2022)
print(f"Updated Car Year: {my_car.get_year()}")

# Trying to access protected attribute (not recommended)
print(f"Fuel Level: {my_car._fuel}")

# Driving the car
my_car.drive(5)
```

```python
'''8. Create a Python class called `BankAccount` with private attributes for the account balance (`__balance`) and
account number (`__account_number`). Provide methods for depositing and withdrawing money.'''
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute for account number
        self.__balance = initial_balance  # Private attribute for account balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount should be greater than zero.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        elif amount > self.__balance:
            print("Insufficient funds.")
        else:
            print("Withdrawal amount should be greater than zero.")

    def get_balance(self):
        return self.__balance  # Getter method to access account balance

    def get_account_number(self):
        return self.__account_number  # Getter method to access account number


# Creating an instance of the BankAccount class
my_account = BankAccount(account_number="123456789", initial_balance=1000)

# Accessing account details using getter methods
print("Account Number:", my_account.get_account_number())
print("Account Balance:", my_account.get_balance())

# Depositing and withdrawing money
my_account.deposit(500)
my_account.withdraw(200)
my_account.withdraw(1500)
```

```python
'''17. Create a Python class for a library system that encapsulates book information, including titles, authors,
and availability status.'''
class Book:
    def __init__(self, title, author):
        self.__title = title  # Private attribute for book title
        self.__author = author  # Private attribute for book author
        self.__available = True  # Private attribute for availability status

    def get_title(self):
        return self.__title  # Getter method for accessing book title

    def get_author(self):
        return self.__author  # Getter method for accessing book author

    def is_available(self):
        return self.__available  # Getter method for checking availability status

    def borrow_book(self):
        if self.__available:
            self.__available = False
            print(f"Book '{self.__title}' by {self.__author} has been borrowed.")
        else:
            print(f"Book '{self.__title}' by {self.__author} is currently not available.")

    def return_book(self):
        if not self.__available:
            self.__available = True
            print(f"Book '{self.__title}' by {self.__author} has been returned.")
        else:
            print(f"Book '{self.__title}' by {self.__author} was not borrowed.")


# Creating instances of the Book class
book1 = Book("Python Crash Course", "Eric Matthes")
book2 = Book("Clean Code", "Robert C. Martin")

# Accessing book information using getter methods
print(f"Book Title: {book1.get_title()}")
print(f"Book Author: {book1.get_author()}")

# Borrowing and returning books
book1.borrow_book()
book1.borrow_book()  # Trying to borrow the same book again
book1.return_book()
book1.return_book()  # Trying to return a book that was not borrowed

book2.borrow_book()
book2.return_book()
```

```python
'''14. Create a Python class hierarchy for a banking system with various account types (e.g., savings, checking, credit card) 
and demonstrate polymorphism by implementing a common `withdraw()` method.'''
class Account:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        pass  # This will be overridden in the subclasses


class SavingsAccount(Account):
    def __init__(self, account_number, interest_rate, balance=0):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawn ${amount} from Savings Account {self.account_number}")
        else:
            print("Insufficient funds")


class CheckingAccount(Account):
    def __init__(self, account_number, balance=0):
        super().__init__(account_number, balance)

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawn ${amount} from Checking Account {self.account_number}")
        else:
            print("Insufficient funds")


class CreditCardAccount(Account):
    def __init__(self, account_number, credit_limit, balance=0):
        super().__init__(account_number, balance)
        self.credit_limit = credit_limit

    def withdraw(self, amount):
        available_credit = self.credit_limit - self.balance
        if available_credit >= amount:
            self.balance += amount
            print(f"Withdrawn ${amount} from Credit Card {self.account_number}")
        else:
            print("Exceeds credit limit")


# Demonstrate polymorphism
def perform_withdraw(account, amount):
    account.withdraw(amount)


# Create instances and perform withdrawals
savings_account = SavingsAccount("SA001", interest_rate=0.05, balance=1000)
checking_account = CheckingAccount("CA001", balance=2000)
credit_card_account = CreditCardAccount("CC001", credit_limit=5000, balance=1000)

perform_withdraw(savings_account, 200)
perform_withdraw(checking_account, 1500)
perform_withdraw(credit_card_account, 2000)
```

```python
'''15. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide
examples using operators like `+` and `*`.'''

'''
Operator overloading in Python allows operators such as +, -, *, /, ==, !=, <, >, etc., to have different meanings for
objects of different classes. It enables defining custom behavior for these operators based on the operands' types.
This concept relates to polymorphism because it allows different classes to provide their own implementation of operators, 
enabling them to behave differently based on the context of the objects involved.
'''
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the addition operator (+)
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    # Overloading the multiplication operator (*)
    def __mul__(self, scalar):
        return Point(self.x * scalar, self.y * scalar)

    def __str__(self):
        return f"({self.x}, {self.y})"


# Creating instances of the Point class
point1 = Point(1, 2)
point2 = Point(3, 4)

# Overloaded addition (+) operator
result_add = point1 + point2
print("Result of addition:", result_add)  

# Overloaded multiplication (*) operator
result_mul = point1 * 3
print("Result of multiplication:", result_mul) 
```

```python
'''17. Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) and implement 
polymorphism through a common `calculate_salary()` method.'''
class Employee:
    def __init__(self, name, role):
        self.name = name
        self.role = role

    def calculate_salary(self):
        return 0  # Base class method, overridden in subclasses


class Manager(Employee):
    def __init__(self, name, salary):
        super().__init__(name, role="Manager")
        self.salary = salary

    def calculate_salary(self):
        return self.salary


class Developer(Employee):
    def __init__(self, name, hourly_rate, hours_worked):
        super().__init__(name, role="Developer")
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked


class Designer(Employee):
    def __init__(self, name, monthly_salary):
        super().__init__(name, role="Designer")
        self.monthly_salary = monthly_salary

    def calculate_salary(self):
        return self.monthly_salary


# Polymorphic function
def total_salary(employees):
    total = 0
    for employee in employees:
        total += employee.calculate_salary()
    return total


# Creating instances of different employee types
manager = Manager("Alice", 5000)
developer = Developer("Bob", 50, 160)
designer = Designer("Charlie", 6000)

# Storing employees in a list
employees = [manager, developer, designer]

# Calculate and print total salary
total = total_salary(employees)
print(f"Total salary for all employees: ${total}")
```

```python
'''6. Create a Python class for a bank account and demonstrate abstraction by hiding the account balance and
providing methods to deposit and withdraw funds.'''
class BankAccount:
    def __init__(self, initial_balance=0):
        self._balance = initial_balance  # Private attribute to store the balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount}. Current balance: ${self._balance}")
        else:
            print("Invalid amount for deposit")

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrawn ${amount}. Current balance: ${self._balance}")
        else:
            print("Insufficient funds or invalid amount for withdrawal")

    # Getter method to access the balance (not necessary but can be used for read-only access)
    def get_balance(self):
        return self._balance


# Example usage
account = BankAccount(1000)  # Initialize account with an initial balance of $1000

print("Initial Balance:", account.get_balance())  # Output: Initial Balance: 1000

account.deposit(500) 
account.withdraw(200)  
account.withdraw(2000)  
account.deposit(-100)

print("Final Balance:", account.get_balance())  
```

```python
'''8. Create a Python class hierarchy for animals and implement abstraction by defining common methods 
(e.g., `eat()`, `sleep()`) in an abstract base class.'''
from abc import ABC, abstractmethod

class Animal(ABC):  # Abstract base class for animals
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def eat(self):
        pass  # Abstract method for eating

    @abstractmethod
    def sleep(self):
        pass  # Abstract method for sleeping

class Dog(Animal):
    def eat(self):
        print(f"{self.name} is eating bones")

    def sleep(self):
        print(f"{self.name} is sleeping in a dog bed")

class Cat(Animal):
    def eat(self):
        print(f"{self.name} is eating fish")

    def sleep(self):
        print(f"{self.name} is sleeping on the couch")

class Snake(Animal):
    def eat(self):
        print(f"{self.name} is eating mice")

    def sleep(self):
        print(f"{self.name} is coiled up and resting")

# Usage
dog = Dog("Buddy")
cat = Cat("Whiskers")
snake = Snake("Slinky")

dog.eat()  
cat.sleep()  
snake.eat() 
```

```python
'''11. Create a Python class for a vehicle system and demonstrate abstraction by defining common methods
(e.g., `start()`, `stop()`) in an abstract base class.'''
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract base class for vehicles
    def __init__(self, make, model):
        self.make = make
        self.model = model

    @abstractmethod
    def start(self):
        pass  # Abstract method for starting the vehicle

    @abstractmethod
    def stop(self):
        pass  # Abstract method for stopping the vehicle

class Car(Vehicle):
    def start(self):
        print(f"{self.make} {self.model} started.")

    def stop(self):
        print(f"{self.make} {self.model} stopped.")

class Motorcycle(Vehicle):
    def start(self):
        print(f"{self.make} {self.model} motorcycle started.")

    def stop(self):
        print(f"{self.make} {self.model} motorcycle stopped.")

# Usage
car = Car("Toyota", "Camry")
motorcycle = Motorcycle("Honda", "CBR")

car.start()  
car.stop()  

motorcycle.start()  
motorcycle.stop() 
```

```python
#12. Describe the use of abstract properties in Python and how they can be employed in abstract classes.
'''
Use of Abstract Properties:
Definition in Abstract Classes: Abstract properties are defined within abstract classes using the @property 
decorator along with @abstractmethod.

Getter and Setter Methods: Abstract properties lack the actual implementation of getter and setter methods. 
They act as a contract, specifying that subclasses must implement these methods.

Enforcing Structure: Abstract properties ensure that subclasses provide their own implementations for accessing or
modifying specific attributes in a consistent manner.'''
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract base class defining abstract properties
    @property
    @abstractmethod
    def brand(self):
        pass  # Abstract property for brand

    @brand.setter
    @abstractmethod
    def brand(self, value):
        pass  # Abstract setter method for brand

    @property
    @abstractmethod
    def color(self):
        pass  # Abstract property for color

    @color.setter
    @abstractmethod
    def color(self, value):
        pass  # Abstract setter method for color

class Car(Vehicle):  # Concrete subclass implementing abstract properties
    def __init__(self):
        self._brand = ""
        self._color = ""

    @property
    def brand(self):  # Getter method for brand
        return self._brand

    @brand.setter
    def brand(self, value):  # Setter method for brand
        self._brand = value

    @property
    def color(self):  # Getter method for color
        return self._color

    @color.setter
    def color(self, value):  # Setter method for color
        self._color = value

# Usage
car = Car()
car.brand = "Toyota"
car.color = "Red"

print("Car Brand:", car.brand)  
print("Car Color:", car.color) 
```

```python
#16. Create a Python class for a computer system, demonstrating abstraction by defining common methods (e.g., `power_on()`, `shutdown()`) in an abstract base class.
from abc import ABC, abstractmethod

class ComputerSystem(ABC):  # Abstract base class for computer systems
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    @abstractmethod
    def power_on(self):
        pass  # Abstract method for powering on the computer system

    @abstractmethod
    def shutdown(self):
        pass  # Abstract method for shutting down the computer system

class Laptop(ComputerSystem):
    def power_on(self):
        print(f"Powering on {self.brand} {self.model} laptop...")

    def shutdown(self):
        print(f"Shutting down {self.brand} {self.model} laptop...")

class Desktop(ComputerSystem):
    def power_on(self):
        print(f"Powering on {self.brand} {self.model} desktop...")

    def shutdown(self):
        print(f"Shutting down {self.brand} {self.model} desktop...")

# Usage
laptop = Laptop("Dell", "Latitude")
desktop = Desktop("HP", "Pavilion")

laptop.power_on() 
laptop.shutdown() 

desktop.power_on()  
desktop.shutdown() 
```

```python
#19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.
from abc import ABC, abstractmethod

class LibrarySystem(ABC):  # Abstract base class for library systems
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def add_book(self, book_title):
        pass  # Abstract method for adding a book to the library system

    @abstractmethod
    def borrow_book(self, book_title, user_id):
        pass  # Abstract method for borrowing a book from the library system

class PublicLibrary(LibrarySystem):
    def __init__(self, name):
        super().__init__(name)
        self.books_available = []

    def add_book(self, book_title):
        self.books_available.append(book_title)
        print(f"Added '{book_title}' to {self.name}")

    def borrow_book(self, book_title, user_id):
        if book_title in self.books_available:
            self.books_available.remove(book_title)
            print(f"User {user_id} borrowed '{book_title}' from {self.name}")
        else:
            print(f"'{book_title}' is not available in {self.name}")

class UniversityLibrary(LibrarySystem):
    def __init__(self, name):
        super().__init__(name)
        self.books_available = []

    def add_book(self, book_title):
        self.books_available.append(book_title)
        print(f"Added '{book_title}' to {self.name}")

    def borrow_book(self, book_title, user_id):
        if book_title in self.books_available:
            self.books_available.remove(book_title)
            print(f"Student {user_id} borrowed '{book_title}' from {self.name}")
        else:
            print(f"'{book_title}' is not available in {self.name}")

# Usage
public_library = PublicLibrary("City Public Library")
university_library = UniversityLibrary("University Central Library")

public_library.add_book("Python Programming")
public_library.add_book("Data Structures and Algorithms")

university_library.add_book("Machine Learning Essentials")
university_library.add_book("Database Systems")

public_library.borrow_book("Python Programming", "User123")
public_library.borrow_book("Machine Learning Essentials", "User456")
```

```python
#5. How can you implement composition in Python classes? Provide examples of using composition to create complex objects.

#ans.In Python, implementing composition involves creating instances of other classes within a class to build complex objects.
class Engine:
    def __init__(self, fuel_type):
        self.fuel_type = fuel_type

    def start(self):
        print(f"Engine started. Fuel type: {self.fuel_type}")


class Wheels:
    def __init__(self, number):
        self.number = number

    def rotate(self):
        print(f"{self.number} wheels are rotating.")


class Transmission:
    def __init__(self, transmission_type):
        self.transmission_type = transmission_type

    def shift_gear(self):
        print(f"Transmission shifted to {self.transmission_type} gear.")


class Car:
    def __init__(self, fuel_type, num_wheels, transmission_type):
        self.engine = Engine(fuel_type)
        self.wheels = Wheels(num_wheels)
        self.transmission = Transmission(transmission_type)

    def start_car(self):
        self.engine.start()
        self.wheels.rotate()
        self.transmission.shift_gear()


# Creating a Car instance using Composition
my_car = Car("Gasoline", 4, "Automatic")

# Using Car methods
my_car.start_car()
```

```python
#6. Create a Python class hierarchy for a music player system, using composition to represent playlists and songs.
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Playing: {self.title} - {self.artist} [{self.duration}]")


class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []  # List to store Song instances

    def add_song(self, song):
        self.songs.append(song)

    def remove_song(self, song):
        if song in self.songs:
            self.songs.remove(song)
        else:
            print(f"{song.title} is not in the playlist.")

    def play_all(self):
        print(f"Playlist: {self.name}")
        for song in self.songs:
            song.play()


# Creating songs
song1 = Song("Song 1", "Artist A", "3:45")
song2 = Song("Song 2", "Artist B", "4:20")
song3 = Song("Song 3", "Artist C", "5:10")

# Creating a playlist and adding songs
my_playlist = Playlist("My Playlist")
my_playlist.add_song(song1)
my_playlist.add_song(song2)
my_playlist.add_song(song3)

# Creating another song and playlist
song4 = Song("Song 4", "Artist D", "2:55")
my_playlist2 = Playlist("Another Playlist")
my_playlist2.add_song(song4)

# Playing songs from playlists
my_playlist.play_all()
my_playlist2.play_all()
```

```python
#8. Create a Python class for a computer system, using composition to represent components like CPU, RAM, and storage devices.
class CPU:
    def __init__(self, brand, cores):
        self.brand = brand
        self.cores = cores

    def get_info(self):
        return f"CPU: {self.brand}, Cores: {self.cores}"


class RAM:
    def __init__(self, size_gb, speed_mhz):
        self.size_gb = size_gb
        self.speed_mhz = speed_mhz

    def get_info(self):
        return f"RAM: {self.size_gb}GB, Speed: {self.speed_mhz}MHz"


class Storage:
    def __init__(self, capacity_gb, type):
        self.capacity_gb = capacity_gb
        self.type = type

    def get_info(self):
        return f"Storage: {self.capacity_gb}GB, Type: {self.type}"


class Computer:
    def __init__(self, cpu, ram, storage):
        self.cpu = cpu
        self.ram = ram
        self.storage = storage

    def get_specs(self):
        specs = []
        specs.append(self.cpu.get_info())
        specs.append(self.ram.get_info())
        specs.append(self.storage.get_info())
        return "\n".join(specs)


# Creating instances of components
my_cpu = CPU("Intel", 8)
my_ram = RAM(16, 3200)
my_storage = Storage(512, "SSD")

# Creating a Computer instance using composition
my_computer = Computer(my_cpu, my_ram, my_storage)

# Getting computer specifications
print(my_computer.get_specs())
```

```python
#10. Create a Python class for a car, using composition to represent components like the engine, wheels, and transmission.
class Engine:
    def __init__(self, fuel_type):
        self.fuel_type = fuel_type

    def start(self):
        return f"Engine started. Fuel type: {self.fuel_type}"


class Wheels:
    def __init__(self, number):
        self.number = number

    def rotate(self):
        return f"{self.number} wheels are rotating."


class Transmission:
    def __init__(self, transmission_type):
        self.transmission_type = transmission_type

    def shift_gear(self):
        return f"Transmission shifted to {self.transmission_type} gear."


class Car:
    def __init__(self, engine, wheels, transmission):
        self.engine = engine
        self.wheels = wheels
        self.transmission = transmission

    def start_car(self):
        return self.engine.start()

    def move(self):
        return f"Car is moving. {self.wheels.rotate()}"

    def change_gear(self):
        return self.transmission.shift_gear()


# Creating instances of components
my_engine = Engine("Gasoline")
my_wheels = Wheels(4)
my_transmission = Transmission("Automatic")

# Creating a Car instance using composition
my_car = Car(my_engine, my_wheels, my_transmission)

# Using Car methods
print(my_car.start_car())
print(my_car.move())
print(my_car.change_gear())
```

```python
#12. Create a Python class for a university course, using composition to represent students, instructors, and course materials.
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def get_student_info(self):
        return f"Student ID: {self.student_id}, Name: {self.name}"


class Instructor:
    def __init__(self, instructor_id, name):
        self.instructor_id = instructor_id
        self.name = name

    def get_instructor_info(self):
        return f"Instructor ID: {self.instructor_id}, Name: {self.name}"


class CourseMaterial:
    def __init__(self, material_name):
        self.material_name = material_name

    def get_material_info(self):
        return f"Course Material: {self.material_name}"


class Course:
    def __init__(self, course_name, instructor, students, course_material):
        self.course_name = course_name
        self.instructor = instructor
        self.students = students
        self.course_material = course_material

    def get_course_info(self):
        info = f"Course: {self.course_name}\n"
        info += f"Instructor: {self.instructor.get_instructor_info()}\n"
        info += "Students:\n"
        for student in self.students:
            info += f"- {student.get_student_info()}\n"
        info += f"Course Material: {self.course_material.get_material_info()}"
        return info


# Creating instances of students, instructor, and course material
student1 = Student(1, "Alice")
student2 = Student(2, "Bob")
instructor = Instructor(101, "Dr. Smith")
material = CourseMaterial("Python Programming")

# Creating a Course instance using composition
students_list = [student1, student2]
python_course = Course("Python Programming", instructor, students_list, material)

# Getting information about the course
print(python_course.get_course_info())
```

```python
#14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.
class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity

    def get_info(self):
        return f"{self.name}: {self.quantity}"


class Dish:
    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients  # List of Ingredient instances

    def get_dish_info(self):
        info = f"Dish: {self.name}\nIngredients:"
        for ingredient in self.ingredients:
            info += f"\n- {ingredient.get_info()}"
        return info


class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes  # List of Dish instances

    def get_menu_info(self):
        info = f"Menu: {self.name}\nDishes:"
        for dish in self.dishes:
            info += f"\n{dish.get_dish_info()}\n"
        return info


# Creating ingredients
ingredient1 = Ingredient("Tomato", "200g")
ingredient2 = Ingredient("Cheese", "100g")
ingredient3 = Ingredient("Bread", "2 slices")

# Creating dishes using ingredients
dish1 = Dish("Margherita Pizza", [ingredient1, ingredient2])
dish2 = Dish("Grilled Cheese Sandwich", [ingredient2, ingredient3])

# Creating a menu with dishes
menu = Menu("Lunch Menu", [dish1, dish2])

# Getting information about the menu
print(menu.get_menu_info())
```

```python
'''16. Create a Python class for a computer game character, using composition to represent attributes like 
weapons, armor, and inventory.'''
class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def attack(self):
        # Logic for attacking
        print(f"Attacking with {self.name} for {self.damage} damage!")


class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def defend(self):
        # Logic for defending
        print(f"Wearing {self.name} for {self.defense} defense!")


class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"Added {item} to inventory.")

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            print(f"Removed {item} from inventory.")
        else:
            print(f"{item} not found in inventory.")

class GameCharacter:
    def __init__(self, name, health):
        self.name = name
        self.health = health
        self.weapon = None
        self.armor = None
        self.inventory = Inventory()

    def equip_weapon(self, weapon):
        self.weapon = weapon
        print(f"{self.name} equipped with {weapon.name}.")

    def equip_armor(self, armor):
        self.armor = armor
        print(f"{self.name} equipped with {armor.name}.")

    def attack(self):
        if self.weapon:
            self.weapon.attack()
        else:
            print("No weapon equipped.")

    def defend(self):
        if self.armor:
            self.armor.defend()
        else:
            print("No armor equipped.")

# Example Usage
sword = Weapon("Sword", 20)
shield = Armor("Shield", 10)

player = GameCharacter("Hero", 100)
player.equip_weapon(sword)
player.equip_armor(shield)
player.attack()
player.defend()

player.inventory.add_item("Health Potion")
player.inventory.remove_item("Mana Potion")
```

```python
#18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.
class Furniture:
    def __init__(self, name):
        self.name = name

    def describe(self):
        return f"This is a {self.name}"


class Appliance:
    def __init__(self, name):
        self.name = name

    def describe(self):
        return f"This is a {self.name}"


class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def describe_room(self):
        room_info = f"{self.name} contains:\n"
        if self.furniture:
            room_info += "Furniture:\n"
            for item in self.furniture:
                room_info += f"- {item.describe()}\n"
        if self.appliances:
            room_info += "Appliances:\n"
            for item in self.appliances:
                room_info += f"- {item.describe()}\n"
        return room_info


class House:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def describe_house(self):
        house_info = "This house has the following rooms:\n"
        for room in self.rooms:
            house_info += room.describe_room()
        return house_info


# Example Usage
living_room = Room("Living Room")
living_room.add_furniture(Furniture("Sofa"))
living_room.add_furniture(Furniture("Coffee Table"))
living_room.add_appliance(Appliance("Television"))

kitchen = Room("Kitchen")
kitchen.add_appliance(Appliance("Oven"))
kitchen.add_appliance(Appliance("Refrigerator"))

my_house = House()
my_house.add_room(living_room)
my_house.add_room(kitchen)

print(my_house.describe_house())
```

```python
#20. Create a Python class for a social media application, using composition to represent users, posts, and comments.
class Comment:
    def __init__(self, text, user):
        self.text = text
        self.user = user

    def __str__(self):
        return f"Comment by {self.user}: {self.text}"


class Post:
    def __init__(self, content, user):
        self.content = content
        self.user = user
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

    def display_comments(self):
        if not self.comments:
            return "No comments yet."
        return "\n".join(str(comment) for comment in self.comments)


class User:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def create_post(self, content):
        post = Post(content, self.username)
        self.posts.append(post)
        return post

    def comment_on_post(self, post, text):
        comment = Comment(text, self.username)
        post.add_comment(comment)


class SocialMediaApp:
    def __init__(self):
        self.users = []

    def create_user(self, username):
        user = User(username)
        self.users.append(user)
        return user


# Example Usage
social_app = SocialMediaApp()

user1 = social_app.create_user("ritu")
user2 = social_app.create_user("rishabh")

post_by_alice = user1.create_post("Hello, everyone!")
user2.comment_on_post(post_by_alice, "Hi ritu!")
user1.comment_on_post(post_by_alice, "Nice to see you, rishabh.")

print(f"Posts by {user1.username}:")
for post in user1.posts:
    print(post.content)
    print(post.display_comments())
```
